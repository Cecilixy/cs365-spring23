```python
import numpy as np
import matplotlib.pyplot as plt
# include if using a Jupyter notebook
%matplotlib inline
import random


def f():
    count = 0
    while True:
        num = random.randint(0,9999)
        if num != 2022:
            count += 1
        else:
            break
    return count

def func(num):
    count = []
    for i in range(num):
        count += [f()]
    return (count)
    
List1 = func(100)
List2 = func(200)
List3 = func(300)
List4 = func(400)
List5 = func(500)
List6 = func(600)
List7 = func(700)
List8 = func(800)
List9 = func(900)
List10 = func(1000)

m1=np.mean(List1)
m2=np.mean(List2)
m3=np.mean(List3)
m4=np.mean(List4)
m5=np.mean(List5)
m6=np.mean(List6)
m7=np.mean(List7)
m8=np.mean(List8)
m9=np.mean(List9)
m10=np.mean(List10)

s1=np.std(List1)
s2=np.std(List2)
s3=np.std(List3)
s4=np.std(List4)
s5=np.std(List5)
s6=np.std(List6)
s7=np.std(List7)
s8=np.std(List8)
s9=np.std(List9)
s10=np.std(List10)

labels = ['100', '200', '300','400','500','600','700','800','900','1000']
x_pos = np.arange(len(labels))
CTEs = [m1, m2, m3,m4,m5,m6,m7,m8,m9,m10]
error = [s1,s2,s3,s4,s5,s6,s7,s8,s9,s10]


fig, ax = plt.subplots()
ax.bar(x_pos, CTEs,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax.set_ylabel('average success')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Problem 2 Strategy 1')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()

plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
# include if using a Jupyter notebook
%matplotlib inline
import random


def f():
    test=[]
    for i in range(10000):
        test+=[i]
    count = 0
    while True:
        num = random.choice(test)
        if num != 2022:
            count += 1
            test.remove(num)
        else:
            break
    return count



def func(num):
    count = []
    for i in range(num):
        count += [f()]
    return (count)
    
List1 = func(100)
List2 = func(200)
List3 = func(300)
List4 = func(400)
List5 = func(500)
List6 = func(600)
List7 = func(700)
List8 = func(800)
List9 = func(900)
List10 = func(1000)

m1=np.mean(List1)
m2=np.mean(List2)
m3=np.mean(List3)
m4=np.mean(List4)
m5=np.mean(List5)
m6=np.mean(List6)
m7=np.mean(List7)
m8=np.mean(List8)
m9=np.mean(List9)
m10=np.mean(List10)

s1=np.std(List1)
s2=np.std(List2)
s3=np.std(List3)
s4=np.std(List4)
s5=np.std(List5)
s6=np.std(List6)
s7=np.std(List7)
s8=np.std(List8)
s9=np.std(List9)
s10=np.std(List10)

labels = ['100', '200', '300','400','500','600','700','800','900','1000']
x_pos = np.arange(len(labels))
CTEs = [m1, m2, m3,m4,m5,m6,m7,m8,m9,m10]
error = [s1,s2,s3,s4,s5,s6,s7,s8,s9,s10]


fig, ax = plt.subplots()
ax.bar(x_pos, CTEs,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax.set_ylabel('average success')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Problem 2 Strategy 2')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.show()
```
